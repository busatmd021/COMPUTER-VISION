## **Fortune 500 Example**

```python
# Line Magic to instruct Jupyter to capture Matplotlib plots and render them in the cell output. 
%matplotlib inline

# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 

# Formatting
sns.set(style="darkgrid")
```

```python
# Load Data
df = pd.read_csv('fortune500.csv')
```

```python
# Inspect the First Rows of Data
df.head()
```

```python
# Inspect Last Few Rows of Data
df.tail()
```

```python
# Rename the Data
df.columns = ['year', 'rank', 'company', 'revenue', 'profit']
```

```python
# Check Data is Complete (expecting 25500)
len(df) 
```

```python
# Check if Data was Imported as Expected
df.dtypes
```

```python
# Check Issue in the Profits Column
non_numberic_profits = df.profit.str.contains('[^0-9.-]')
df.loc[non_numberic_profits].head()
```

```python
# Fix Issue with the Profit Data
set(df.profit[non_numberic_profits])
len(df.profit[non_numberic_profits])
```

```python
# Decide Whether or not to Remove the Data
bin_sizes, _, _ = plt.hist(df.year[non_numberic_profits], bins=range(1955, 2006))
```

```python
# Deemed Acceptable to Remove the Data
df = df.loc[~non_numberic_profits]
df.profit = df.profit.apply(pd.to_numeric)

# Check it Worked
len(df)
df.dtypes
```

```python
# Now the Data is Cleaned it can be Used
group_by_year = df.loc[:, ['year', 'revenue', 'profit']].groupby('year')
avgs = group_by_year.mean()
x = avgs.index
y1 = avgs.profit
def plot(x, y, ax, title, y_label):
    ax.set_title(title)
    ax.set_ylabel(y_label)
    ax.plot(x, y)
    ax.margins(x=0, y=0)
```

```python
# Plot the Profit Data
fig, ax = plt.subplots()
plot(x, y1, ax, 'Increase in mean Fortune 500 company profits from 1955 to 2005', 'Profit (millions)')
```

```python
# Plot the Revenue Data
y2 = avgs.revenue
fig, ax = plt.subplots()
plot(x, y2, ax, 'Increase in mean Fortune 500 company revenues from 1955 to 2005', 'Revenue (millions)')
```

```python
# Superimpose the Plots ontop of one another
def plot_with_std(x, y, stds, ax, title, y_label):
    ax.fill_between(x, y - stds, y + stds, alpha=0.2)
    plot(x, y, ax, title, y_label)
fig, (ax1, ax2) = plt.subplots(ncols=2)
title = 'Increase in mean and std Fortune 500 company %s from 1955 to 2005'
stds1 = group_by_year.std().profit.values
stds2 = group_by_year.std().revenue.values
plot_with_std(x, y1.values, stds1, ax1, title % 'profits', 'Profit (millions)')
plot_with_std(x, y2.values, stds2, ax2, title % 'revenues', 'Revenue (millions)')
fig.set_size_inches(14, 4)
fig.tight_layout()
```

### **Learning Widgets**

```python
import ipywidgets as widgets
from IPython.display import display

def update_plot(year_range):
    start_year, end_year = year_range
    mask = (x >= start_year) & (x <= end_year)

    fig, ax = plt.subplots(figsize=(10, 6))
    plot(x[mask], y1[mask], ax, f'Increase in mean Fortune 500 company profits from {start_year} to {end_year}', 'Profit (millions)')
    plt.show()

year_range_slider = widgets.IntRangeSlider(
    value=[1955, 2005],
    min=1955,
    max=2005,
    step=1,
    description='Year range:',
    continuous_update=False
)

widgets.interact(update_plot, year_range=year_range_slider)
```
